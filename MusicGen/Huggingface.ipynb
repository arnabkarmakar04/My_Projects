```python
from transformers import AutoProcessor, MusicgenForConditionalGeneration
import torch
import scipy.io.wavfile
from IPython.display import Audio
```

```python
processor = AutoProcessor.from_pretrained("facebook/musicgen-small")
model = MusicgenForConditionalGeneration.from_pretrained("facebook/musicgen-small")
```

```python
prompts = ("Bollywood Beat","A fast-paced Bollywood dance beat with electronic drums and sitar")
```

```python
inputs = processor(text=prompts, padding=True, return_tensors="pt")
```

```python
audio_values = model.generate(
    **inputs,
    do_sample=True,
    guidance_scale=4,
    max_new_tokens=1024
)
```

```python
sampling_rate = model.config.audio_encoder.sampling_rate
audio_array = audio_values[0, 0].numpy()
scipy.io.wavfile.write("musicgen_out.wav", rate=sampling_rate, data=audio_array)
print("Music Generated !!!")
```

```python
from IPython.display import Audio
Audio(audio_values[0].numpy(), rate=sampling_rate)
```

```python
duration_seconds = len(audio_array) / sampling_rate
print(f"Duration: {duration_seconds:.2f} seconds")
```
